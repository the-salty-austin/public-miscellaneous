```python
# Get Data from Ceiba
url = 'https://ceiba.ntu.edu.tw/course/88ca22/content/training.txt'
content = requests.get(url).content
content = content.decode('utf-8')

data = {}
content = content.replace(' \r', '').replace(' ', ',').strip().split('\n')

for c in content:
    if c.startswith('13'):
        c = c[:-1]  # direct downloading has a weird " " at the end
    c = [int(x) for x in c.split(',')]
    data[ c[0] ] = c[1:]
```

### Split then Validate

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.svm import SVC
```

```python
X_train, X_test, y_train, y_test = train_test_split(trainX, trainY, test_size=0.1, random_state=46)
```

```python
svmlin_model = SVC(kernel='linear', C=1)
svmlin_model.fit(X_train, y_train)

predicted_results = []
predicted_results.extend(svmlin_model.predict(X_test))
print(predicted_results)
```

```python
print(metrics.confusion_matrix(y_test, predicted_results))
```

- error rate

```python
mat = metrics.confusion_matrix(y_test, predicted_results)
dim = len( mat )
err = 0
for i in range(dim):
  for j in range(dim):
    if i <= j: continue
    if mat[i][j] > 0:
      err += mat[i][j]

err / len(X_test), err
```

## What about dimension reduction then SVM?

### PCA Reduction

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=128)
```

```python
cls_vecs = np.array( [x[0] for x in embeddings] )
cls_vecs_reduce = pca.fit_transform(cls_vecs)
```

```python
cls_vecs_reduce.shape
```

```python
IS_OUT = []
for i in range(1, 1096):
    IS_OUT.append(True)

trainX_reduce = []
trainY_reduce = []

for cls, vals in data.items():
  for val in vals:
    trainX_reduce.append( cls_vecs_reduce[val-1] )
    trainY_reduce.append( cls )
    IS_OUT[val-1] = False

trainX_reduce = np.array(trainX_reduce)
trainY_reduce = np.array(trainY_reduce)
# ===

outX_reduce = []
out_indices = []

for i in range(1, 1096):
  if IS_OUT[i-1]:
    outX_reduce.append( cls_vecs_reduce[i-1] )
    out_indices.append(i)

outX_reduce = np.array(outX_reduce)
```

```python
trainX_reduce.shape, outX_reduce.shape
```

```python
svmlin_red_model = SVC(kernel='linear', C=1.02)
svmlin_red_model.fit( trainX_reduce, trainY )
```

```python
out_predicted_red_results = []
out_predicted_red_results.extend(svmlin_red_model.predict( outX_reduce ) )
```

```python
import csv

with open('SVM_red.csv', 'w', newline='') as f:
    # create the csv writer
    writer = csv.writer(f)
    writer.writerow(['Id', 'Value'])
    for x, y in zip(out_indices, out_predicted_red_results):
        writer.writerow([x, int(y)])
```
